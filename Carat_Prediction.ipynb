```python
# importing libraries 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# preparing the test data and train data
df = pd.read_csv('diamonds.csv')
df = df.dropna()
x_train, x_test, y_train, y_test = train_test_split(df['price'],df['carat'],random_state=0,train_size=0.63)
x_train = np.array(x_train).reshape(-1,1)
x_test = np.array(x_test).reshape(-1,1)
y_train = np.array(y_train).reshape(-1,1)
y_test = np.array(y_test).reshape(-1,1)
```

```python
# traning the model with test data
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
# ploting graph for x_train and y_train
plt.scatter(x_train,y_train,color='r')
plt.title("Price vs Carat")
plt.xlabel("Price")
plt.ylabel("Carat")
plt.show()
```

```python
# ploting graph for xtrain and prediction made by model 
y_pred = lr.predict(x_test)
plt.plot(x_test,y_pred,color='b')
plt.scatter(x_train,y_train,color='r')
plt.title("Price vs Carat")
plt.xlabel("Price")
plt.ylabel("Carat")
plt.show()
```
